*Objective*:
What is k means neighbours algorithm.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X_train=np.random.normal(0,1,100)
y_train=np.array([1]*50+[0]*50)
X_test=np.random.normal(0,1,10)

df=pd.DataFrame(np.vstack((X_train,y_train)).transpose(), columns=['X','Y'])
df.head()
```

```python
df.Y.value_counts()
```

```python
K=5

def find_kNN(x, X_train, y_train, K):
    dist=abs(x-X_train)

    # print('1st',dist.shape)
    dist=np.vstack((dist,y_train)).transpose()
    # print('2nd',dist.shape)
    # print(np.argsort(dist[:,0]))
    labels=dist[np.argsort(dist[:,0])][:K][:,1]
    # This line sorts all points by their distance to 'x', selects the K closest ones,
    # and then extracts their associated labels.
    # remember np.argsort(dist[:,0] , these return the indexes to sort

    count=sum(labels)
    if (count>int(K/2)):
        test_label=1
    else:
        test_label=0

    return test_label
```

```python
y_test=list(map(lambda x: find_kNN(x, X_train, y_train, K), X_test))
y_test
```

```python
plt.scatter(X_train,y_train,label='train')
plt.scatter(X_test,y_test,label='val')
plt.legend() # legend is for labels

plt.show()
```
